## Broadcasting

### The term broadcasting describes how NumPy treats arrays with different shapes during arithmetic operations.
### The smaller array is broadcast across the larger array so that they have compatible shapes.

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)
print(a)
print(b)
print(a+b)
```

```python
# diff shape
a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)
print(a)
print(b)
print(a+b)
```

# RULES

```python
# if x=m and Y=n,operation wii be take place
A1=np.arange(8).reshape(2,4)
A2=np.arange(8,16).reshape(2,4)
print (A1+A2)
```

```python
# if x=1 and Y=n,operation wii be take place(same dimension)
A3=np.arange(3).reshape(1,3)
A4=np.arange(12).reshape(4,3)
print (A3+A4)
```

```python
# if x=1 and Y!=n,operation wii be not take place
A5=np.arange(3).reshape(1,3)
A6=np.arange(16).reshape(4,4)
print (A5+A6)
#operands could not be broadcast together with shape (1,3) and (4,4)
```

```python
# if x=1 and n=1 then y=m,operation wii be take place
A7=np.arange(3).reshape(1,3)
A8=np.arange(3).reshape(3,1)
print (A7+A8)
```

```python
# if x=1 and Y=1,operation wii be take place
A9=np.arange(1).reshape(1,1)
A10=np.arange(20).reshape(4,5)
print (A9+A10)
```

```python
# if they are different dimensions 
A11=np.arange(4)
A12=np.arange(20).reshape(5,4)
print (A11+A12)
```

```python
a = np.arange(10)
np.sin(a)
```

```python
def sigmoid(array):
  return 1/(1 + np.exp(-(array)))
```

```python
a = np.arange(100)
sigmoid(a)
```

```python
# mean squared error
actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)
```

```python
def mse(actual,predicted):
  return np.mean((actual - predicted)**2)
mse(actual,predicted)
```

```python
# binary cross entropy
np.mean((actual - predicted)**2)
```
